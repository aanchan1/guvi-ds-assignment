```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/text processing.csv')
df.head(2)
```

```python
df1=df[['text']]
df1.head()
```

lower case

```python
df1['text_lower']=df1['text'].str.lower()
df1
```

upper case

```python
df1['text_upper']=df1['text'].str.upper()
df1
```

removing punctuation

```python
import string
df1.drop(["text_lower", "text_upper"], axis=1, inplace=True)

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df1["text_wo_punct"] = df1["text"].apply(lambda text: remove_punctuation(text))
df1
```

stemming

```python
from nltk.stem.porter import PorterStemmer

df1.drop(["text_wo_punct", "text_wo_stop"], axis=1, inplace=True) 

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df1["text_stemmed"] = df1["text"].apply(lambda text: stem_words(text))
df1.head()
```
